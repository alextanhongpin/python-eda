```python
from collections import namedtuple

Station = namedtuple("Station", "codes name")
stations = []
name_by_code = {}
for codes, name in data:
    name = name.strip()
    valid_codes = [code for code in codes.split("\xa0") if len(code.strip()) > 1]
    for code in valid_codes:
        name_by_code[code] = name
    stations.append(Station(valid_codes, name))
```

```python
def split_alphanum(key):
    n = int("".join([k for k in key if k.isdigit()]) or 0)
    s = "".join([k for k in key if k.isalpha()])
    return [s, n]


sorted_codes = sorted(name_by_code, key=lambda k: split_alphanum(k))
sorted_codes
```

```python
from itertools import groupby

import networkx as nx


def extract_numeric(s):
    n = "".join(c for c in s if c.isdigit())
    return 0 if n == "" else int(n)


excluded_codes = set()
G = nx.DiGraph()
for code, codes in groupby(sorted_codes, key=lambda k: k[:2]):
    all_codes = list(codes)
    # Ignore if only 1 station. It must not some special station.
    if len(all_codes) == 1:
        print("Ignoring", all_codes)
        excluded_codes.add(all_codes[0])
        continue
    a, b, *rest = all_codes
    # Ignore if the station number is not continuous.
    if abs(extract_numeric(a) - extract_numeric(b)) > 2:
        print("Ignoring", all_codes)
        for code in all_codes:
            excluded_codes.add(code)
        continue
    for a, b in zip(all_codes, all_codes[1:]):
        G.add_edge(a, b, weight=1)
        G.add_edge(b, a, weight=1)
excluded_codes
```

```python
for station in stations:
    if excluded_codes & set(station.codes):
        continue
    match station.codes:
        case (a,):
            pass
        case a, b:
            G.add_edge(a, b, weight=0)
            G.add_edge(b, a, weight=0)
        case a, b, c:
            G.add_edge(a, b, weight=0)
            G.add_edge(b, a, weight=0)
            G.add_edge(a, c, weight=0)
            G.add_edge(c, a, weight=0)
            G.add_edge(b, c, weight=0)
            G.add_edge(c, b, weight=0)
```

```python
# https://networkx.org/documentation/stable/reference/algorithms/approximation.html
# https://stackoverflow.com/questions/14527815/how-to-fix-the-start-and-end-points-in-travelling-salesmen-problem
from itertools import combinations

from networkx.algorithms import approximation

H = nx.DiGraph()

start = "DT1"
nodes = ["CC16", "NS19"]
end = "CC20"
dummy = "DUMMY"

H.add_edge(dummy, start, weight=0)
H.add_edge(end, dummy, weight=0)

for start, end in combinations([start, end, *nodes], 2):
    weight = nx.shortest_path_length(G, start, end)
    print(start, end, nx.shortest_path(G, start, end), weight)
    H.add_edge(start, end, weight=weight)
    H.add_edge(end, start, weight=weight)


paths = approximation.traveling_salesman_problem(H)
paths
```
